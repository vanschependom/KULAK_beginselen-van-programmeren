In deze sessie beginnen we met **iteraties**: het is handing dat we een computer beslissingen kunnen laten nemen, maar we willen vaak taken automatiseren waar men veel snelle beslissinen na elkaar moet nemen.

# While

Een while-lus gebruiken we als we iets willen uitvoeren **zolang een bepaalde conditie is voldaan**. Hieronder zien jullie enkele voorbeelden van zo'n situaties:

```python
# Tel getallen op uit een lijst zolang resultaat kleiner is dan 10:
som = 0
while(som < 10):
    #som += x is een kortere schrijfwijze voor som = som + x
    som += int(input("geef een getal om op te tellen "))
print("meer dan 10!")
print(f"De som is: {som}")
```

```python
#Vraag getallen zolang het getal geen 42 is:
getal = 0
while(getal != 42):
    getal = int(input("Wat is het antwoord op het leven, het universum en de rest? "))
print("correct")
```

# For

Een for-lus **itereert voor bepaalde waarden van een variabele**. Een for loop heeft dus 2 componenten: een variabele, die elke lus "leidt" en een verzameling, die bepaalt wat de waarde van die variabele wordt.


Belangrijk is om in te zien dat een for lus een centrale variabele heeft ( VOOR deze waarde van x, doe ...). Dit heeft als grote kracht dat we weten iets weten over de lus waar we in zitten. Kijk eens naar de voorbeelden voor while en probeer in te zien dat het daar geen belang heeft de hoeveelste lus bezig is.

Om te weten wanneer je een for-lus moet gebruiken (of als je dit omdraait: als je dit leest moet je zeker een for-lus overwegen) moet je het probleem als volgt kunnen verwoorden: "voor elke ... in ... moet ik ...". Dit kan dan op enkele manieren concreten worden opgevat: 
- Je wil iets 15, 100, ... keer doen, en je weet vooraf hoeveel keer. Hier doe je iets VOOR elke waarde tussen 0 en 15, 100, ...
- Je wil iets VOOR elk object in een verzameling doen. Bijvoorbeeld VOOR elk getal x in mijn verzameling [1,2,3,4], hier is x die centrale variabele.
- ...

Let wel dat we later dieper op verzamelingen zullen ingaan. Nu is het gewoon belangrijk dat je snapt dat [1,2.0,'3',"vier"] een verzameling is met 4 elementen: 1, 2.0, '3' en "vier".

Hieronder staan weer enkele voorbeelden waar we een for-lus gebruiken:

```python
# Voor elk getal van 1 tot 20, geef me het product met 5.
# Het gegeven "van 1 tot 20" wordt in python omgezet met range(1,20). Het commando range(x,y) geeft dus een verzameling
# die alle getallen bevat vanaf 1 tot 20. Lees dus "voor elke waarde in [1,2,...,18,19]"
for i in range(1,20):
    print(i*5)
```

```python
# Voor elke waarde in ["pokémon","digimon","kardemom"] print de laatste 3 letters met "ster, heel speciaal voor jou" er achter:
for woord in ["pokémon","digimon","kardemom"]:
    print(f"een {woord[-3:]}ster, heel speciaal voor jou")
```

```python
# Bereken de som voor elke 4e getal tussen 1 en 20. (dit is dus 1,5,9,...)
som = 0
for getal in range(1,20,4):
    som += getal
print(som)
```

# modules

Een laatste aspect dat we hier willen uitleggen is modules. Modules in python zijn manieren om code en commando's aan te bieden die niet tot de kern van python behoren. De math-module geeft ons toegang tot verschillende functies uit de wiskunde, die niet tot de basis behoren van python. De random-module geeft ons toegang tot willekeurige getallenLet wel dat deze modules standaard worden meegeleverd, maar we beschouwen "if, else, while, input, int, ..." als elementairer dan "vierkantswortel, afronden, willekeurig getal genereren, ..." voor een programmeer taal.

Belangrijk is hier dan je de commando's kan gebruiken en lezen, maar we zullen altijd wel hints geven als die functionaliteit in een module zit, je zal nooit verwacht worden om te weten of iets wel of niet in die module aanwezig is.

```python
#importeer de math module, zodat de interpreter weet dat je deze wil gebruiken
import math

#gebruik de vierkantswortel-functie (square root)
print(math.sqrt(4))

#rond een getal af naar beneden
print(math.floor(2.1))
print(math.floor(4.9))

#rond een getal af naar boven
print(math.ceil(5.1))
print(math.ceil(3.6))

# verhef een getal tot een macht
print(math.pow(7,2))

#geef de sinus van een hoek (gegeven in radialen)
print(math.sin(1.3))

#geef een mooi wiskundig getal voor cirkels, dat toch wel ergens in de top-100 mooiste wiskundige getallen zit
print(math.pi)
```

```python
#importeer de module, zodat de interpreter weet dat je deze wil gebruiken
import random

#gebruik de integer-functie, deze genereert een geheel getal van a tot en met b uniform verdeeld
print(random.randint(1,4))

#genereer een getal tussen 0 en 1 dat uniform is verdeeld
print(random.random())

#print de waarde 'groen' met kans 1/2 en 'geel' of 'rood' met kans 1/4
if random.random() > 0.5:
    print("groen")
else:
    if random.random() > 0.5:
        print("geel")
    else:
        print("rood")

#selecteer willekeurig een object uit een gegeven verzameling
print(random.choice([1,2.0,'3',"vier"]))
```

# Oefening 3:

Aan de hand van een Monte Carlo simulatie, kunnen we $\pi$ benaderen. Hiervoor genereren we $N$ aantal willekeurige punten in het vlak [[-1,1],[-1,1]]. Door de verhouding te nemen van het aantal punten die in de ingesloten cirkel vallen en het totaal aantal gegenereerde punten, benaderen we $\pi/4$. Kijk naar de verhouding van de oppervlakten om jezelf hiervan te overtuigen.

Maak nu een programma dat $\pi$ berekend tot op een gegeven precisie.




```python
import math
import random

# Vraag de precisie aan de gebruiker (geef een niet te klein getal in).
gevraagde_precisie = max(0.0001,float(input("Geef de gewenste precisie in: ")))

# initialiseer de nodige variabelen:
pi_geschat = 0
aantalBinnenCirkel = 0
totaal = 0

while abs(math.pi-pi_geschat) > gevraagde_precisie :

    totaal +=1
    
    # de random functie geeft een willekeurig getal terug tussen [0,1]
    x = (random.random() * 2) - 1
    y = (random.random() * 2) - 1
    
    # bereken de straal van de pool-coordinaten
    r = math.sqrt(x**2+y**2)
    
    # indien het gegenereerde punt binnen de cirkel valt pas je bepaalde variabelen aan
    if abs(r) <= 1 :
        aantalBinnenCirkel+=1

    if aantalBinnenCirkel > 100:
        pi_geschat = (aantalBinnenCirkel/totaal) * 4
    
print(f"pi werd berekend to op een precisie van {gevraagde_precisie}: {pi_geschat}")
```
